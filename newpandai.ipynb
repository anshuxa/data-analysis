```python
import pandas as pd
from pandasai import Agent
import os
from pandasai.llm.google_gemini import GoogleGemini
from pandasai import SmartDataframe
```

```python
from google.colab import userdata

GOOGLE_API_KEY=userdata.get('GOOGLE_API_KEY')
```

```python


data_dict = {
    "country": [
        "Delhi",
        "Mumbai",
        "Kolkata",
        "Chennai",
        "Jaipur",
        "Lucknow",
        "Pune",
        "Bengaluru",
        "Amritsar",
        "Agra",

    ],
    "annual tax collected": [
        19294482072,
        28916155672,
        24112550372,
        34358173362,
        17454337886,
        11812051350,
        16074023894,
        14909678554,
        43807565410,
        146318441864

    ],
    "happiness_index": [9.94, 7.16, 6.35, 8.07, 6.98, 6.1, 4.23, 8.22, 6.87, 3.36],}
df = pd.DataFrame(data_dict)
df.head()
```

```python
llm = GoogleGemini(api_key=GOOGLE_API_KEY)
```

```python
pandas_ai = SmartDataframe(df, config={"llm": llm, "verbose": True})
```

```python
pandas_ai.chat('which is the happiest country')
```
